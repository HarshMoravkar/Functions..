Q1.What is the difference between a function and a method in Python?

  -In Python, both functions and methods are used to perform specific tasks, but they differ in how they are defined and used.

  A function is a standalone block of code that is defined using the def keyword. It is not associated with any object or class and can be called directly by its name. For example, a function might look like this:

  def add(a, b):

  return a + b

  On the other hand, a method is a function that is defined inside a class and is associated with an object. It automatically takes the instance (usually named self) as its first parameter.

  For example:

  class Calculator:

  def add(self, a, b):

  return a + b

Q3.What are the different ways to define and call a function in Python?

  -In Python, functions can be defined and called in multiple ways depending on the purpose and use case.

  1.Standard Function Definition (using def)

  You define a function using the def keyword, followed by the function name and parameters.

  def greet(name):

  return f"Hello, {name}!"

  Calling the function:

  print(greet("Alice"))

  2.Function with Default Arguments

  You can provide default values to parameters.

  def greet(name="Guest"):

  return f"Hello, {name}!"

  greet()        

  greet("Alice")  

  3.Variable-Length Arguments

  You can define functions that accept any number of arguments using *args and **kwargs.

  def add_all(*numbers):

  return sum(numbers)

  def print_info(**details):

  for key, value in details.items():

   print(f"{key}: {value}")

  add_all(1, 2, 3, 4)

  print_info(name="Alice", age=25)

  4.Lambda Functions (Anonymous Functions)

  Short, single-expression functions can be created using the lambda keyword.

  square = lambda x: x * x

  print(square(5))  # Output: 25

  5.Function as an Object / Passing Functions

  Functions can be passed as arguments or returned from other functions.

  def shout(text):

  return text.upper()

  def call_function(func, value):

  return func(value)

  call_function(shout, "hello")  # Output: "HELLO"

  6.Recursive Functions

  Functions that call themselves.

  def factorial(n):

   if n == 0:

   return 1

   return n * factorial(n - 1)

  factorial(5)  # Output: 120

Q5.What are iterators in Python and how do they differ from iterables?

  -In Python, iterables and iterators are two related concepts used for looping over data, but they are not the same.

  Iterable:
  An iterable is any Python object capable of returning its elements one at a time. Examples include lists, tuples, strings, sets, and dictionaries.

  To be an iterable, an object must implement the __iter__() method, which returns an iterator.

  Example:

  my_list = [1, 2, 3]

  for item in my_list:  

  print(item)

   Iterator:
  An iterator is an object that keeps track of the current position while iterating. It implements two methods:

  __iter__() – returns the iterator object itself.

  __next__() – returns the next item in the sequence and raises StopIteration when there are no more items.

  You can create an iterator from an iterable using the iter() function.

  Example:

  my_list = [1, 2, 3]
  it = iter(my_list)  

  print(next(it))  
  print(next(it))  
  print(next(it))
  
Q6.Explain the concept of generators in Python and how they are defined?

  -In Python, a generator is a type of iterable, like a list or tuple, but instead of storing all values in memory at once, it generates values on the fly as you iterate over it. This makes generators more memory-efficient, especially for large datasets or infinite sequences.

  Generators are defined using a function but instead of the return statement, they use the yield keyword to produce a series of values.

  How to Define a Generator:
  You define a generator just like a normal function, but use yield instead of return.

  Example:

  def count_up_to(n):

  i = 1

  while i <= n:

  yield i

  i += 1

  When this function is called, it returns a generator object:

  gen = count_up_to(3)

  print(next(gen))  

  print(next(gen))  

  print(next(gen))  

Q7.What are the advantages of using generators over regular functions?

  -Generators offer several advantages over regular functions, especially when dealing with large datasets or streaming data. The key benefits are:

  1.Memory Efficiency
  Generators yield one item at a time and do not store the entire sequence in memory. This makes them ideal for handling large data or infinite sequences.

  2.Lazy Evaluation
  Generators evaluate values on the fly, meaning values are only computed when needed. This reduces CPU and memory usage.

  3.Infinite Sequences
  With generators, you can create infinite sequences because they don’t require storing all values.

  4.Better Performance
  For many tasks, generators can be faster because they avoid overhead from creating and storing large collections.

  5.Clean and Simple Code
  Generators simplify the code required for custom iteration logic, especially compared to writing a class with __iter__() and __next__() methods.

  6.Supports Pipelining
  Generators can be chained together to process data step-by-step, similar to Unix pipes.

Q9.Explain the purpose and usage of the `map()` function in Python?

  -The map() function in Python is used to apply a given function to every item of an iterable (such as a list, tuple, etc.) and return a map object (which is an iterator) with the results.

  Purpose:
  To transform or process each element in an iterable by applying a function.

  It helps avoid explicit loops for simple operations, making the code cleaner and often more efficient.

  Syntax:

  map(function, iterable, ...)

  function: The function to apply to each item.

  iterable: One or more iterables whose elements are passed to the function.

  Example:

  def square(x):

  return x * x

  numbers = [1, 2, 3, 4]

  squares = map(square, numbers)  # Returns a map object

  print(list(squares))  # Output: [1, 4, 9, 16]

  Using map() with Multiple Iterables:

  If multiple iterables are passed, the function must take that many arguments and map() applies the function to the items from each iterable in parallel.

  numbers1 = [1, 2, 3]

  numbers2 = [4, 5, 6]

  result = map(lambda x, y: x + y, numbers1, numbers2)

  print(list(result))  # Output: [5, 7, 9]

Q10.What is the difference between `map()`, `reduce()`, and `filter()`functions in Python?

  -map(), reduce(), and filter() are built-in (or from the functools module for reduce()) functions in Python used for functional programming to process iterables. They serve different purposes:

  1.map()

  Purpose: Applies a given function to each item of an iterable and returns a map object with the results.

  Usage: Transform or modify every element.

  Example:

  numbers = [1, 2, 3]

  squares = map(lambda x: x**2, numbers)

  print(list(squares))  # Output: [1, 4, 9]

  2.filter()

  Purpose: Filters items from an iterable based on a function that returns True or False.

  Usage: Select elements that satisfy a condition.

  Example:

  numbers = [1, 2, 3, 4, 5]

  evens = filter(lambda x: x % 2 == 0, numbers)

  print(list(evens))  # Output: [2, 4]

  3.reduce()

  Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.

  Usage: Perform rolling computation like summing, multiplying, or concatenating.

  Note: reduce() is in the functools module, so it needs to be imported.

  Example:

  from functools import reduce

  numbers = [1, 2, 3, 4]

  total = reduce(lambda x, y: x + y, numbers)
  
  print(total)  # Output: 10

```python
#Q1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

my_list = [47, 11, 42, 13, 8, 6]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)

```

```python
#Q3.Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    squared = []
    for num in numbers:
        squared.append(num ** 2)
    return squared
input_list = [2, 3, 4, 5]
result = square_numbers(input_list)
print("Squared list:", result)
```

```python
#Q4.Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False  # 0 and 1 are not prime
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
for number in range(1, 201):
    if is_prime(number):
        print(number, "is a prime number")
```

```python
#Q5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
fib = FibonacciIterator(10)

for number in fib:
    print(number)
```

```python
#Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp
for value in powers_of_two(5):
    print(value)
```

```python
#Q7.Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Removes trailing newline characters
for line in read_file_line_by_line('A.txt'):
    print(line)

```

```python
#Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple?

my_list = [(1, 3), (4, 1), (2, 2), (5, 0)]
sorted_list = sorted(my_list, key=lambda x: x[1])
print(sorted_list)
```

```python
#Q9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit?

def c_to_f(celsius):
    return (celsius * 9/5) + 32
celsius_temps = [0, 10, 20, 30, 40]
fahrenheit_temps = list(map(c_to_f, celsius_temps))
print(fahrenheit_temps)
```

```python
#Q10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda char: char not in vowels, input_string))
text = "Hello, how are you?"
result = remove_vowels(text)
print(result)
```
